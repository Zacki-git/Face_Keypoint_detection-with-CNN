```python
import re
import matplotlib.pyplot as plt
```

```python
def get_loss_from_log(data):
    metadata_file = data
    with open(metadata_file) as f:
        lines = f.readlines()
    train_list = []
    valid_list = []
    for i in range(len(lines)):
        # print(i)
        loss = int(re.findall(r"\d+", lines[i])[-2])

        if i % 3 == 1:
            train_list.append(loss)
        elif i % 3 == 2:
            valid_list.append(loss)
    return train_list, valid_list
```

```python
def show_diff(list1, list2, optimizerName,start=None):
    plt.figure(figsize=(30, 20), dpi=80)
    plt.ylim = max(list1) * 1.1
    plt.xlim = len(list2) * 1.1

    plt.plot(list1[start:], label="train_loss", c="r", marker=".")
    plt.plot(list2[start:], label="valid_loss", c="g", marker=".")

    plt.legend(loc="upper right", fontsize="30")

    plt.title("Comparison " + str(optimizerName) + "_Loss" + "\n" +
              "from epoch {} to epoch {}".format(start, len(list1) - 1), fontsize="30")

    plt.show()
```

```python
data1 = 'log_info_round1.txt'
train_list1, valid_list1 = get_loss_from_log(data1)
data2 = 'log_info_round2.txt'
train_list2, valid_list2 = get_loss_from_log(data2)
data3 = 'log_info_round3.txt'
train_list3, valid_list3 = get_loss_from_log(data3)
data4 = 'log_info_round4.txt'
train_list4, valid_list4 = get_loss_from_log(data4)
data5 = 'log_info_round5.txt'
train_list5, valid_list5 = get_loss_from_log(data5)
data6 = 'log_info_round6.txt'
train_list6, valid_list6 = get_loss_from_log(data6)
```

```python
show_diff(train_list1, valid_list1,"adam")
```

```python
show_diff(train_list2, valid_list2,"adam")
```

```python
show_diff(train_list3, valid_list3,"adam")
```

```python
show_diff(train_list4, valid_list4,"adam")
```

```python
show_diff(train_list5, valid_list5,"sgd")
```

```python
show_diff(train_list6, valid_list6,"sgd")
```
